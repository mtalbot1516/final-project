```python
import pandas as pd
import itertools
```

```python
li = list(range(1,25))
df = pd.DataFrame()
for x in li:
    file_path =  'raw_files/sold{}.csv'.format(x)
    frame = pd.read_csv(file_path)
    df = df.append(frame)
```
